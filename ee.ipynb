```python
import numpy as np
import matplotlib.pyplot as plt

import modules_py.functions as f
import modules_py.functions_vec as fv

plt.style.use('publication23.mplstyle')
```

```python
# pf = 1.28 # pf 0.873, 1.12, 1.28 | n 0.1, 0.2, 0.3

plt.plot(fv.k, -fv.eq_pnn(fv.k, 0, 0.41).real, label=r'$n = %.2f n_0$' % f.n(0.41), ls='-.', color='black')
plt.plot(fv.k, -fv.eq_pnn(fv.k, 0, 0.873).real, label=r'$n = %.1f n_0$' % f.n(0.873), ls='-', color='red')
plt.plot(fv.k, -fv.eq_pnn(fv.k, 0, 1.26).real, label=r'$n = %.1f n_0$' % f.n(1.26), ls='--', color='green')
plt.plot(fv.k, -fv.eq_pnn(fv.k, 0, 1.38).real, label=r'$n = %.1f n_0$' % f.n(1.38), ls=':', color='blue')

plt.axhline(y=0, ls=':', lw=1, color='black')

plt.xlim(0, 5)
plt.ylim(-1, 5)

plt.xlabel(r'$k$ $[m_\pi]$')
plt.ylabel(r'$\tilde \omega ^ 2(k)$ $[m_\pi^2]$')
# plt.title(r'eq$(\omega = 0, k)$')
plt.legend(frameon=False, prop={'size': 15})

plt.savefig('figures/misc/ee_eq_pnn_w0.pdf', dpi=300, format='pdf', bbox_inches='tight')
# plt.savefig('figures/misc/ee_eq_pnn_w0.jpg', dpi=300, format='JPG', bbox_inches='tight')
plt.show()
```
